```python
import pandas as pd

import xmltodict
from urllib.request import urlopen
import json
```

```python
res_dict = []
for j in range(1,3):
    url = "https://itunes.apple.com/kr/rss/customerreviews/page=%i/id=597333937/sortby=mostrecent/xml?urlDesc=/customerreviews/id=597333937/sortBy=mostRecent/xml" % j
    response = urlopen(url).read()

    xml = xmltodict.parse(response)
    XmlToJson = json.loads(json.dumps(xml))

    for i in range(len(XmlToJson['feed']['entry'])):
        res_dict.append({
            'date' : XmlToJson['feed']['entry'][i]['updated'][:10],
            'ratings' : int(XmlToJson['feed']['entry'][i]['im:rating']),
            'helpful' : int(XmlToJson['feed']['entry'][i]['im:voteSum']),
#             'DISLIKE' : int(XmlToJson['feed']['entry'][i]['im:voteCount']) - int(XmlToJson['feed']['entry'][i]['im:voteSum']),
#             'TITLE' : XmlToJson['feed']['entry'][i]['title'],
            'comment' : XmlToJson['feed']['entry'][i]['content'][0]['#text']
        })
```

```python
res_df = pd.DataFrame(res_dict)
res_df['date'] = pd.to_datetime(res_df['date'], format="%Y-%m-%d")

res_df.to_csv('리브온_Appstore.csv', encoding='utf-8-sig', index=False)
res_df
```
